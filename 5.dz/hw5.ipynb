```python
from matrix import Matrix
import numpy as np
import matplotlib.pyplot as plt
import integral
%pylab inline
```

```python
A = Matrix.create("a1.txt")
print(A)
A_inv = A.inv()
```

```python
A = Matrix.create("a2.txt")
print(A)
A_inv = A.inv()
print(A_inv)
print(A * A_inv)
print(A_inv * A)
```

```python
A = Matrix.create("a3.txt")
A = np.array(A.data)
B = np.array([[0],[0]])
x0 = np.array([[0.0],[2.0]])

T = 0.1
t_max = 5
x = integral.trapz(x0, A, B, T, t_max, trace=True, step=5)
```

```python
k = int(t_max / T+1)
t = np.linspace(0, t_max, k)
plt.plot(t, x[:, 0])
plt.plot(t, x[:, 1]);
```
